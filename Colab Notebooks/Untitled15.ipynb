```python
import pandas as pd

# Read the dataset
df = pd.read_csv('elcsales.csv')

# 1. What are the column names in the dataset?
column_names = df.columns.tolist()
print("Column names:", column_names)
print(df)

# 2. How many rows are there in the dataset?
num_rows = df.shape[0]
print("Number of rows:", num_rows)

# 3. How many columns are there in the dataset?
num_columns = df.shape[1]
print("Number of columns:", num_columns)

# 4. What is the total sales amount?
total_sales = df['Sale_amt'].sum()
print("Total sales amount:", total_sales)

# 5. What is the average unit price?
avg_unit_price = df[' Unit_price '].mean()
print("Average unit price:", avg_unit_price)

# 6. What is the maximum number of units sold?
max_units_sold = df['Units'].max()
print("Maximum units sold:", max_units_sold)

# 7. Which region had the highest sales?
region_highest_sales = df.groupby('Region')['Sale_amt'].sum().idxmax()
print("Region with the highest sales:", region_highest_sales)

# 8. How many different items are there in the dataset?
num_items = df['Item'].nunique()
print("Number of different items:", num_items)

# 9. What is the total sales amount for each manager?
sales_by_manager = df.groupby('Manager')['Sale_amt'].sum()
print("Sales amount by manager:")
print(sales_by_manager)

# 10. What is the total sales amount for each salesman?
sales_by_salesman = df.groupby('SalesMan')['Sale_amt'].sum()
print("Sales amount by salesman:")
print(sales_by_salesman)

# 11. What is the average unit price for each item?
avg_unit_price_by_item = df.groupby('Item')['Unit_price'].mean()
print("Average unit price by item:")
print(avg_unit_price_by_item)

# 12. What is the total sales amount for each month?
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
df['Month'] = df['OrderDate'].dt.month
sales_by_month = df.groupby('Month')['Sale_amt'].sum()
print("Sales amount by month:")
print(sales_by_month)

# 13. What is the total sales amount for each year?
df['Year'] = df['OrderDate'].dt.year
sales_by_year = df.groupby('Year')['Sale_amt'].sum()
print("Sales amount by year:")
print(sales_by_year)

# 14. What is the average unit price for each region?
avg_unit_price_by_region = df.groupby('Region')['Unit_price'].mean()
print("Average unit price by region:")
print(avg_unit_price_by_region)

# 15. What is the total sales amount for each item in the East region?
sales_by_item_east = df[df['Region'] == 'East'].groupby('Item')['Sale_amt'].sum()
print("Sales amount by item in the East region:")
print(sales_by_item_east)

# 16. What is the total sales amount for each item in the Central region?
sales_by_item_central = df[df['Region'] == 'Central'].groupby('Item')['Sale_amt'].sum()
print("Sales amount by item in the Central region:")
print(sales_by_item_central)

# 17. What is the average unit price for each manager in the West region?
avg_unit_price_manager_west = df[df['Region'] == 'West'].groupby('Manager')['Unit_price'].mean()
print("Average unit price by manager in the West region:")
print(avg_unit_price_manager_west)

# 18. What is the total sales amount for each salesman in the East region?
sales_by_salesman_east = df[df['Region'] == 'East'].groupby('SalesMan')['Sale_amt'].sum()
print("Sales amount by salesman in the East region:")
print(sales_by_salesman_east)

# 19. What is the total sales amount for each month in the Central region?
sales_by_month_central = df[df['Region'] == 'Central'].groupby('Month')['Sale_amt'].sum()
print("Sales amount by month in the Central region:")
print(sales_by_month_central)

# 20. What is the average unit price for each item in the East region?
avg_unit_price_item_east = df[df['Region'] == 'East'].groupby('Item')['Unit_price'].mean()
print("Average unit price by item in the East region:")
print(avg_unit_price_item_east)
```
