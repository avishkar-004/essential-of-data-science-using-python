```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime

# Step 1: Read the CSV file using pandas
data= pd.read_csv('/content/356_assesment 5.csv')  # Replace 'your_file_name.csv' with your actual file name

data['DataUsage'] = [float(usage[:-2]) for usage in data['DataUsage']]
# Bar Chart - Monthly Bill
plt.bar(data['CustomerID'], data['MonthlyBill'])
plt.xlabel('Customer ID')
plt.ylabel('Monthly Bill')
plt.title('Monthly Bill by Customer')
plt.show()
print("\n")
# Pie Chart - Data Usage Distribution
plt.pie(data['DataUsage'], labels=data['Plan'], autopct='%1.1f%%')
plt.title('Data Usage Distribution')
plt.show()
print("\n")
# Line Chart - Minutes Used Over Time
plt.plot(data['CustomerID'], data['MinutesUsed'])
plt.xlabel('Customer ID')
plt.ylabel('Minutes Used')
plt.title('Minutes Used Over Time')
plt.show()
print("\n")
# Scatter Plot - Data Usage vs. Monthly Bill
plt.scatter(data['DataUsage'], data['MonthlyBill'])
plt.xlabel('Data Usage')
plt.ylabel('Monthly Bill')
plt.title('Data Usage vs. Monthly Bill')
plt.show()
print("\n")
# Bar Chart - City-wise Data Usage
plt.bar(data['City'], data['DataUsage'])
plt.xlabel('City')
plt.ylabel('Data Usage')
plt.title('Data Usage by City')
plt.show()
print("\n")
# Pie Chart - Plan Distribution
plt.pie(pd.Series(data['Plan']).value_counts(), labels=pd.Series(data['Plan']).unique(), autopct='%1.1f%%')
plt.title('Plan Distribution')
plt.show()
print("\n")
# Line Chart - Texts Sent Over Time
plt.plot(data['CustomerID'], data['TextsSent'])
plt.xlabel('Customer ID')
plt.ylabel('Texts Sent')
plt.title('Texts Sent Over Time')
plt.show()
print("\n")
# Stacked Bar Chart - Plan Distribution by City
df_grouped = pd.DataFrame(data).groupby(['City', 'Plan']).size().unstack()
df_grouped.plot(kind='bar', stacked=True)
plt.xlabel('City')
plt.ylabel('Count of Customers')
plt.title('Plan Distribution by City')
plt.legend(title='Plan')
plt.show()
print("\n")
# Scatter Plot - Minutes Used vs. Texts Sent
plt.scatter(data['MinutesUsed'], data['TextsSent'])
plt.xlabel('Minutes Used')
plt.ylabel('Texts Sent')
plt.title('Minutes Used vs. Texts Sent')
plt.show()
print("\n")
# Bar Chart - Top 5 Customers by Monthly Bill
top_customers = pd.DataFrame(data).nlargest(5, 'MonthlyBill')
plt.bar(top_customers['CustomerID'], top_customers['MonthlyBill'])
plt.xlabel('Customer ID')
plt.ylabel('Monthly Bill')
plt.title('Top 5 Customers by Monthly Bill')
plt.show()
```
