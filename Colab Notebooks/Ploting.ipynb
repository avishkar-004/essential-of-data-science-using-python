```python
x = np.linspace(0,10,100)
fig = plt.figure()
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
fig.savefig('graph1.png')
```

```python
plt.plot(x,np.sin(x-0),color='blue')
plt.plot(x,np.sin(x-1),color='g')
plt.plot(x,np.sin(x-2),color='0.75')
plt.plot(x,np.sin(x-3),color='#FFDD44')
plt.plot(x,np.sin(x-3),color=(1.0,0.2,0.3))
```

```python
plt.plot(x,np.sin(x-0),color='blue',linestyle = 'solid')
plt.plot(x,np.sin(x-1),color='blue',linestyle = 'dashed')
plt.plot(x,np.sin(x-2),color='blue',linestyle = 'dashdot')
```

```python
import matplotlib.pyplot as plt

# Sample data
categories = ['Day 1', 'Day 2', 'Day 3', 'Day 4']
values = [25, 40, 30, 50]

# Create a bar plot
plt.bar(categories, values)

# Customize the plot
plt.title("Bar Plot Example")
plt.xlabel("Categories")
plt.ylabel("Values")

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
categories = ['Category 1', 'Category 2', 'Category 3', 'Category 4']
values = [25, 40, 30, 50]

# Create a bar plot
plt.barh(categories, values)

# Customize the plot
plt.title("Bar Plot Example")
plt.xlabel("Categories")
plt.ylabel("Values")

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# create data
x = np.arange(5)
y1 = [34, 56, 12, 89, 67]
y2 = [12, 56, 78, 45, 90]
width = 0.40

# plot data in grouped manner of bar type
plt.bar(x-0.2, y1, width)
plt.bar(x+0.2, y2, width)

```
