```python

import logging
from telegram import Update
from telegram.ext import Updater, CommandHandler, CallbackContext

# Set up logging
logging.basicConfig(format='%(asctime)s - %(name)s - %(levelname)s - %(message)s', level=logging.INFO)

# Define the start command handler
def start(update: Update, context: CallbackContext):
    menu = '''
        Welcome to XYZ Canteen!

        Menu:
        1. Chicken Parmesan - $12.99
        2. Margherita Pizza - $10.99
        3. Grilled Salmon - $15.99
        4. Spaghetti Bolognese - $9.99
        5. Caesar Salad - $7.99
        6. Beef Burger - $11.99
        7. Vegetarian Pizza - $9.99
        8. Shrimp Scampi - $14.99
        9. Mushroom Risotto - $8.99
        10. Chocolate Brownie Sundae - $6.99

        Beverages:
        - Soft Drinks (Coke, Pepsi, Sprite) - $2.49
        - Iced Tea (Sweetened/Unsweetened) - $2.49
        - Fresh Orange Juice - $3.99
        - Cappuccino - $3.49
        - Bottled Water - $1.99

        Enjoy your meal at XYZ Canteen!
    '''
    context.bot.send_message(chat_id=update.effective_chat.id, text=menu)

# Define the help command handler
def help(update: Update, context: CallbackContext):
    help_message = '''
        /start --> Welcome to the Canteen
/help --> This particular message
/contact --> Contact information
    '''
    context.bot.send_message(chat_id=update.effective_chat.id, text=help_message)

# Define the contact command handler
def contact(update: Update, context: CallbackContext):
    contact_message = '''
        Email : xyz_canteen@gmail.com
Phone no : +91889034XXXX

NOTE : In Case the bot does not work kindly direct yourself to t.me/canteen_service_bot
    '''
    context.bot.send_message(chat_id=update.effective_chat.id, text=contact_message)

def main():
    # Create an updater object
    updater = Updater("6057010092:AAHxk-k2lFJnvew-MD4dPhQbdNqPoYmNdRc", use_context=True)

    # Get the dispatcher to register handlers
    dispatcher = updater.dispatcher

    # Register the command handlers
    dispatcher.add_handler(CommandHandler("start", start))
    dispatcher.add_handler(CommandHandler("help", help))
    dispatcher.add_handler(CommandHandler("contact", contact))

    # Start the bot
    updater.start_polling()

    # Run the bot until you press Ctrl-C
    updater.idle()

if __name__ == '__main__':
    main()
```
