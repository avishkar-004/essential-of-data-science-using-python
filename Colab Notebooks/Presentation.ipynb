## Python Sets


```python
thisset = {"apple", "banana", "cherry"}
print(thisset)
```

```python
thisset = set(("apple", "banana", "cherry"))
print(thisset)
```

Duplicates are not allowed

```python
thisset = {"apple", "banana", "cherry", "apple"}

print(thisset)
```

```python
thisset = {"apple", "banana", "cherry"}

for x in thisset:
  print(x)
```

```python
thisset = {"apple", "banana", "cherry"}

print("banana" in thisset)
```

```python
thisset = {"apple", "banana", "cherry"}

thisset.add("orange")

print(thisset)
```

```python
thisset = {"apple", "banana", "cherry"}
tropical = {"pineapple", "mango", "papaya"}

thisset.update(tropical)

print(thisset)
```

```python
thisset = {"apple", "banana", "cherry"}
mylist = ["kiwi", "orange"]

thisset.update(mylist)

print(thisset)
```

```python
thisset = {"apple", "banana", "cherry"}

thisset.remove("banana")

print(thisset)
```

```python
thisset = {"apple", "banana", "cherry"}

thisset.discard("banana")

print(thisset)
```

```python
thisset = {"apple", "banana", "cherry"}

x = thisset.pop()

print(x) #removed item

print(thisset) #the set after removal
```

```python
thisset = {"apple", "banana", "cherry"}

thisset.clear()

print(thisset)
```

```python
thisset = {"apple", "banana", "cherry"}

del thisset

print(thisset)
```

```python
thisset = {"apple", "banana", "cherry"}

for x in thisset:
  print(x)
```

```python
set1 = {"a", "b" , "c"}
set2 = {1, 2, 3}

set3 = set1.union(set2)
print(set3)
```

```python
set1 = {"a", "b" , "c"}
set2 = {1, 2, 3}

set1.update(set2)
print(set1)
```

```python
x = {"apple", "banana", "cherry"}
y = {"google", "microsoft", "apple"}

x.intersection_update(y)

print(x)
```

```python
x = {"apple", "banana", "cherry"}
y = {"google", "microsoft", "apple"}

z = x.intersection(y)

print(z)
```

```python
x = {"apple", "banana", "cherry"}
y = {"google", "microsoft", "apple"}

x.symmetric_difference_update(y)

print(x)
```

```python
x = {"apple", "banana", "cherry"}
y = {"google", "microsoft", "apple"}

z = x.symmetric_difference(y)

print(z)
```
