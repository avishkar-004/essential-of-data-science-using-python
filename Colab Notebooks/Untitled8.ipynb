```python
price=list()
age=list()
fuel=list()



f=open("toyota1.csv","r")
data=f.readline()
while(True):

  data=f.readline()
  if not data:
    break;
  print(data)
  temp=data.split(",")
  price.append(int(temp[0]))
  age.append(int(temp[1]))
  fuel.append(temp[3])
```

```python
print(price,"\n\n")

print("Highest price of car is",max(price))
```

```python
print(price)
print(age,"\n\n")

print("Age of car having lowest price is",age[price.index(min(price))])
```

```python
print("Lowest price of car is",min(price))
```

```python
print("Average price of car is",sum(price)/len(price))
```

```python
print(fuel,"\n\n")
frequency ={}
for car in fuel:
  if car in frequency:
    frequency[car]+=1
  else:
    frequency[car]=1
print(frequency,"\n\n")
```
