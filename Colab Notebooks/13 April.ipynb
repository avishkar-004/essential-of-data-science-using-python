```python
78745#even and odd no 
List1 = []
Even = []
Odd = []
 
n = int(input("How many elements in list :- "))
 
for i in range(n):
    value = int(input())
    List1.append(value)
 
for j in range(n):
    if(List1[j] % 2 == 0):
        Even.append(List1[j])
    else:
        Odd.append(List1[j])
 
print("\n Element in Even List is : ", Even)
print("\n Element in Odd List is : ", Odd)
```

```python
# seperate in list of 3,5,10
List1 = []
Div3 = []
Div5 = []
Div10 = []
n = int(input("How many elements in list :- "))
 
for i in range(n):
    value = int(input("Please enter the Value of %d Element :- " %i))
    if(value % 3 == 0):
        Div3.append(value)
    if(value % 5 == 0):
        Div5.append(value)
    if(value % 10 == 0):
        Div10.append(value)
 
print("\n Element divisible by 3 is : ", Div3)
print("\n Element divisible by 5 is : ", Div5)
print("\n Element divisible by 10 is : ", Div10)
```

```python
# BREAK STATEMENT
for i in range (1,10):
   if (i>5):
    break
   print(i)
```

```python
# CONTINUE STATEMENT
for i in range (1,10):
   if (i==5):
    continue
   print(i)
   
```
