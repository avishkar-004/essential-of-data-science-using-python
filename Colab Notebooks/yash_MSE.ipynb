```python
sal=list()
sal_usd=list()
location=list()
size=list()


f=open("company1.csv","r")
data=f.readline()
while(True):

  data=f.readline()
  if not data:
    break;
  print(data)
  temp=data.split(",")
  sal.append(int(temp[4]))
  sal_usd.append(int(temp[5]))
  location.append(temp[7])
  size.append(temp[-1])
```

```python
print(sal,"\n\n")

print(min(sal),"is lowest salary in ruppes\n\n")
```

```python
print(sal_usd,"\n\n")

print(max(sal_usd),"USD is highest salary in doller\n\n")
```

```python
print(location,"\n\n")

frequency ={}
for job in location:
  if job in frequency:
    frequency[job]+=1
  else:
    frequency[job]=1
print(frequency,"\n\n")

print(frequency["CA"],"employees having job location as Canada")
```

```python
print("Job location wise Count:\n",frequency)
```

```python
print(size,"\n\n")

frequency ={}
for job in size:
  if job in frequency:
    frequency[job]+=1
  else:
    frequency[job]=1
print(frequency,"\n\n")

print(frequency["L\n"],"employees are working on large scale company")
```
