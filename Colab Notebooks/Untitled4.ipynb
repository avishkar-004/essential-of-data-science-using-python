```python
import csv


f1 = open ("emp.csv","r")
f2 = open ("sal.csv","r")
f3 = open ("emp_sal.csv","w")

d1=list(csv.reader(f1,delimiter=','))
d2=list(csv.reader(f2,delimiter=','))

print("\n\nFile1 Conttents:",d1)
print("\n\nFile1 Conttents:",d2)
d3=[]
for i in range (len(d1)):
  d3.append(d1[i]+d2[i])


print(d3)
cw=csv.writer(f3)
cw.writerows(d3)

f1.close()
f2.close()
f3.close()

f=open("emp_sal.csv","r")

contents=f.read()

lines=contents.split("\n")
print("\n\n\n\n\n")
sal=[]

for l in lines :
  words=l.split(",")
  print(words)
  sal.append(words[-1])

  

print("Maximum salary",max(sal))
print("Minimum salary",min(sal))
print("sum of salary")

```
