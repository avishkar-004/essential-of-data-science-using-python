```python
from pandas import DataFrame
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
Data = {'x':
[0.893,1.115,1.050,0.896,1.018,0.926,0.891,0.921,0.926,1.159,1.139,1.126,1.100,1.130,1.059,1.125,1.069,0.998,0.921,0.879,0.935,1.015,0.998,1.126,1.122,1.119],

"y":[0.12,0.26,0.37,0.46,0.45,0.56,0.67,0.69,0.86,0.71,0.64,0.73,0.01,0.95,0.85,0.64,0.91,0.82,0.78,0.33,0.41,0.61,0.32,0.21,0.08,0.98]}
df=DataFrame(Data,columns=['x','y'])
print(df)
```

```python
km=KMeans(n_clusters=3).fit(df)
centroids=km.cluster_centers_
print(centroids)
plt.xlabel("Incident Light")
plt.ylabel("Reflected Light")
plt.scatter(df['x'],df['y'],c=km.labels_.astype(float),s=60,alpha=1)
plt.scatter(centroids[:,0],centroids[:,1],c='red',s=190)
```
